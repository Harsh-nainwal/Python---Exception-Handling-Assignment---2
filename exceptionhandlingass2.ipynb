Note: Here Exception class refers to the base class for all the exceptions.

Ans - By deriving our custom exception class from the `Exception` class, we ensure that our custom exception inherits the essential behaviors and functionalities provided by the `Exception` class. This includes the ability to be raised, caught, and handled like other built-in exceptions, making it consistent with the existing exception hierarchy in Python.

Q2. Write a python program to print Python Exception Hierarchy.

```python
def classtree(cls, indent=0):
    print ('.' * indent, cls.__name__)
    for subcls in cls.__subclasses__():
        classtree(subcls, indent + 3)

classtree(BaseException)
```

Q4. Why LookupError class is used? Explain with an example KeyError and IndexError.

ANS - The LookupError class in Python is the base class for exceptions that occur when a lookup or indexing operation fails. It provides a common parent class for various lookup-related exceptions.

KeyError: Raised when a key is not found in a dictionary or mapping.

```python
my_dict = {"a": 1, "b": 2, "c": 3}

try:
    value = my_dict["d"]
    print("Value:", value)
except KeyError:
    print("Error: Key not found")
```

IndexError: Raised when trying to access an element in a sequence using an invalid index.

```python
l = [1,2,3,4,5,6,7]
try :
    print(l[10])
except IndexError as e:
    print(e)
```

Q5. Explain ImportError. What is ModuleNotFoundError?

Ans - ImportError: This exception is raised when an import statement fails to find or load a module.

ModuleNotFoundError: This exception is a subclass of ImportError and is raised when a module or package cannot be found.

```python
try : 
    import my_module
except ModuleNotFoundError as e:
    print(e)
```

```python
try:
    import non_existent_module
except ImportError:
    print("Error: Module not found or unable to import")
```
